```python
housing = pd.read_csv('files/housing.csv')
housing
```

#### Train Test Split

```python
# train test split
housing_train,housing_test=train_test_split(housing,test_size=0.2,stratify=housing['ocean_proximity'],random_state=42)
```

```python
# target and feature dataset
x_train=housing.drop(['median_house_value'],axis=1)
y_train=housing['median_house_value']
x_train
```

```python
y_train
```

#### model training

```python
lin_reg=Pipeline([
    ("preprocessing",preprocessings),
    ("linearregression",LinearRegression())
])
```

```python
lin_reg.fit(x_train,y_train)
```
