```python
import pandas as pd
import numpy as np
from preprocessing import preprocessings

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

from sklearn.model_selection import cross_val_score

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline

from sklearn.metrics import mean_squared_error
```

#### Steps
1. Train test split data
2. Explore the data
3. Preprocess the data
4. Select the model
5. Train and Evaluate the training set

```python
housing = pd.read_csv('files/housing.csv')
housing
```

#### Train Test Split

```python
# train test split
housing_train,housing_test=train_test_split(housing,test_size=0.2,stratify=housing['ocean_proximity'],random_state=42)
```

```python
housing_train.shape
```

```python
# target and feature dataset
x_train=housing.drop(['median_house_value'],axis=1)
y_train=housing['median_house_value']
x_train
```

```python
y_train
```

#### model training

```python
lin_reg=Pipeline([
    ("preprocessing",preprocessings),
    ("linearregression",LinearRegression())
])
```

```python
lin_reg.fit(x_train,y_train)
```

#### rough idea about model information

```python
lin_reg.predict(x_train[:5])
```

#### Evaluate the model on training set

```python
predictions=lin_reg.predict(x_train)
predictions
```

```python
lin_mse = mean_squared_error(y_train, predictions)
lin_rmse=np.sqrt(lin_mse)
lin_rmse
```

##### Model Underfitting
1. data not informative
2. model not powerful

### using more complex model

```python
# from sklearn.tree import DecisionTreeRegressor
```

```python
tree=Pipeline([
    ("preprocessing",preprocessings),
    ("decisionTree",DecisionTreeRegressor())
])
tree.fit(x_train,y_train)
```

#### performance measure of tree

```python
pred_tree=tree.predict(x_train)
```

```python
tree_mse=mean_squared_error(y_train,pred_tree)
tree_rmse=np.sqrt(tree_mse)
tree_rmse
```

#### Model Overfit

```python
housing_test
```

#### Better Evaluation using cross-validation


```python
# from sklearn.model_selection import cross_val_score
```

```python
tree_rmses = -cross_val_score(tree,x_train,y_train,scoring="neg_root_mean_squared_error",cv=10)
tree_rmses
```

```python
#  training_set--> 0.0
# validation set --> 65000
# Model Overfitting

# regularize the model
# reduce the noice in dataset

np.mean(tree_rmses)
```

### Looking for more complex/powerful Model

```python
# from sklearn.ensemble import RandomForestRegressor
```

```python
random_forest=Pipeline([
    ("preprocessing",preprocessings),
    ("randomforest",RandomForestRegressor(random_state=42))
])
```

```python
random_forest.fit(x_train,y_train)
```

```python
rnd_pred=random_forest.predict(x_train)
rnd_pred
```

```python
y_train
```

```python
forest_mse = mean_squared_error(y_train, rnd_pred)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

### Model Evaluation

### cross_val_score

```python
forest_cv_score=cross_val_score(random_forest,x_train,y_train,cv=10,scoring="neg_root_mean_squared_error")
```
